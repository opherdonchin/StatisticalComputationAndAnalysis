```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
import io
import scipy.stats as stats
import seaborn as sns
```

```python
#mean estimator distribution

#population is normal distribution with mean of 100 and standard deviation of 10
#sample 100 datapoints each time
#repeat 1000 times to create estimator distribution
samples1000 = np.random.normal(100, 10, size = (100,1000))
AllMeans = np.mean(samples1000, axis = 0)


#plot results
plt.hist(AllMeans)
plt.xlabel('Sample Means', fontsize = 14)
plt.ylabel('Count', fontsize = 14)
plt.title('Mean Estimator Distribution', fontsize = 14)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.savefig("Mean Estimator Distribution.png", bbox_inches='tight')
files.download("Mean Estimator Distribution.png")

plt.figure()
plt.hist(AllMeans)

plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.plot([100, 100], [0, 300], color = 'pink', lw = 3, label = 'True Mean = 100')
plt.plot([np.mean(AllMeans), np.mean(AllMeans)], [0, 300], color = 'red', lw = 3, label = f'Mean = {np.mean(AllMeans)}')

plt.plot([np.mean(AllMeans) - np.std(AllMeans), np.mean(AllMeans) - np.std(AllMeans)], [0, 300], color = 'violet', lw = 3, label = f'std = {np.std(AllMeans)}')
plt.plot([np.mean(AllMeans) + np.std(AllMeans), np.mean(AllMeans) + np.std(AllMeans)], [0, 300], color = 'violet', lw = 3)
plt.xlabel('Sample Means', fontsize = 14)
plt.ylabel('Count', fontsize = 14)
plt.title('Mean Estimator Distribution', fontsize = 14)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.legend(fontsize = 12, loc = 'upper left')

plt.savefig("Mean Estimator Distribution2.png", bbox_inches='tight')
files.download("Mean Estimator Distribution2.png")
```

```python
#bias and variance

#population is normal distribution with mean of 100 and standard deviation of 10
#sample 100 datapoints each time
#repeat 1000 times to create estimator distribution
samples1000 = np.random.normal(100, 10, size = (100,1000))
AllMeans = np.mean(samples1000, axis = 0)


#plot results
plt.hist(AllMeans)
plt.xlabel('Sample Means', fontsize = 14)
plt.ylabel('Count', fontsize = 14)
plt.title('Mean Estimator Distribution', fontsize = 14)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.plot([np.mean(AllMeans), np.mean(AllMeans)], [0, 300], color = 'red', lw = 3, label = 'True Mean')
plt.legend(fontsize = 12)
plt.xlim([86, 114])

plt.savefig("small_bias.png", bbox_inches='tight')
files.download("small_bias.png")

plt.figure()
plt.hist(AllMeans)
plt.xlabel('Sample Means', fontsize = 14)
plt.ylabel('Count', fontsize = 14)
plt.title('Mean Estimator Distribution', fontsize = 14)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.plot([np.mean(AllMeans) + 10 , np.mean(AllMeans) + 10], [0, 300], color = 'red', lw = 3, label = 'True Mean')
plt.legend(fontsize = 12)
plt.xlim([86, 114])

plt.savefig("large_bias.png", bbox_inches='tight')
files.download("large_bias.png")


#variance
samples10 = np.random.normal(100, 10, size = (5,1000))
AllMeans2 = np.mean(samples10, axis = 0)


#plot results
plt.figure()
plt.hist(AllMeans2)
plt.xlabel('Sample Means', fontsize = 14)
plt.ylabel('Count', fontsize = 14)
plt.title('Mean Estimator Distribution', fontsize = 14)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.plot([np.mean(AllMeans), np.mean(AllMeans)], [0, 300], color = 'red', lw = 3, label = 'True Mean')
plt.legend(fontsize = 12)

plt.savefig("large_var.png", bbox_inches='tight')
files.download("large_var.png")

plt.figure()
plt.hist(AllMeans2)
plt.xlabel('Sample Means', fontsize = 14)
plt.ylabel('Count', fontsize = 14)
plt.title('Mean Estimator Distribution', fontsize = 14)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.plot([np.mean(AllMeans) + 10, np.mean(AllMeans) + 10], [0, 300], color = 'red', lw = 3, label = 'True Mean')
plt.legend(fontsize = 12)

plt.savefig("large_var_and_bias.png", bbox_inches='tight')
files.download("large_var_and_bias.png")
```

```python
#standard distribution estimator distribution
samples = np.random.normal(100, 10, size = (5,1000))
AllStdsBiased = np.std(samples, axis = 0)
AllStdsunBiased = np.std(samples, axis = 0, ddof = 1)


#plot results

plt.figure()
plt.hist(AllStdsBiased)
plt.xlabel('Sample Stds', fontsize = 14)
plt.ylabel('Count', fontsize = 14)
plt.title('Biased Standard Deviation Estimator Distribution', fontsize = 14)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.plot([10, 10], [0, 300], color = 'red', lw = 3, label = 'True Standard Deviation = 10')
plt.plot([np.mean(AllStdsBiased), np.mean(AllStdsBiased)], [0, 300], color = 'violet', lw = 3, label = f'Mean of Estimator Distribution = {np.round(np.mean(AllStdsBiased), 2)}')
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.legend(fontsize = 12)
plt.xlim([0, 20])

plt.savefig("Biased.png", bbox_inches='tight')
files.download("Biased.png")

plt.figure()
plt.hist(AllStdsunBiased)
plt.xlabel('Sample Stds', fontsize = 14)
plt.ylabel('Count', fontsize = 14)
plt.title('Unbiased Standard Deviation Estimator Distribution', fontsize = 14)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.plot([10, 10], [0, 300], color = 'red', lw = 3, label = 'True Standard Deviation = 10')
plt.plot([np.mean(AllStdsunBiased), np.mean(AllStdsunBiased)], [0, 300], color = 'violet', lw = 3, label = f'Mean of Estimator Distribution = {np.round(np.mean(AllStdsunBiased), 2)}')

plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.legend(fontsize = 12)
plt.xlim([0, 20])

plt.savefig("UnBiased.png", bbox_inches='tight')
files.download("UnBiased.png")

#larger sample size
samples = np.random.normal(100, 10, size = (100,1000))
AllStdsBiased = np.std(samples, axis = 0)
AllStdsunBiased = np.std(samples, axis = 0, ddof = 1)


#plot results

plt.figure()
plt.hist(AllStdsBiased)
plt.xlabel('Sample Stds', fontsize = 14)
plt.ylabel('Count', fontsize = 14)
plt.title('Biased Standard Deviation Estimator Distribution', fontsize = 14)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.plot([10, 10], [0, 300], color = 'red', lw = 3, label = 'True Standard Deviation = 10')
plt.plot([np.mean(AllStdsBiased), np.mean(AllStdsBiased)], [0, 300], color = 'violet', lw = 3, label = f'Mean of Estimator Distribution = {np.round(np.mean(AllStdsBiased), 2)}')
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.legend(fontsize = 12)
plt.xlim([0, 20])

plt.savefig("Biased100.png", bbox_inches='tight')
files.download("Biased100.png")

plt.figure()
plt.hist(AllStdsunBiased)
plt.xlabel('Sample Stds', fontsize = 14)
plt.ylabel('Count', fontsize = 14)
plt.title('Unbiased Standard Deviation Estimator Distribution', fontsize = 14)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.plot([10, 10], [0, 300], color = 'red', lw = 3, label = 'True Standard Deviation = 10')
plt.plot([np.mean(AllStdsunBiased), np.mean(AllStdsunBiased)], [0, 300], color = 'violet', lw = 3, label = f'Mean of Estimator Distribution = {np.round(np.mean(AllStdsunBiased), 2)}')

plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.legend(fontsize = 12)
plt.xlim([0, 20])

plt.savefig("UnBiased100.png", bbox_inches='tight')
files.download("UnBiased100.png")
```
